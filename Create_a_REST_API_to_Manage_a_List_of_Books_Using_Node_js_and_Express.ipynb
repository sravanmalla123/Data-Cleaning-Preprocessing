## Implement POST Add Book

### Subtask:
Create an Express POST endpoint (`/books`) that adds a new book to the in-memory data store. Generate a unique ID for the new book and return the newly created book.


### Implement POST Add Book

To create an Express POST endpoint (`/books`) that adds a new book to the in-memory data store, follow these steps:

1.  **Open your `server.js` (or `app.js`) file.**
2.  **Ensure you have `app.use(express.json());` configured** at the top of your `server.js` file, after `const app = express();`, to parse JSON request bodies. If it's not there, add it. (This was already included in the last full `server.js` code block, but it's good to re-emphasize).
3.  **Add the following JavaScript code** after your existing GET endpoints. This code defines a new POST route that listens for requests to `/books`:

    ```javascript
    app.post('/books', (req, res) => {
        const { title, author } = req.body;
        if (!title || !author) {
            return res.status(400).send('Title and author are required');
        }
        // Simple unique ID generation (incrementing current length for simplicity)
        const newBook = {
            id: String(books.length > 0 ? Math.max(...books.map(b => parseInt(b.id))) + 1 : 1),
            title,
            author
        };
        books.push(newBook);
        res.status(201).json(newBook);
    });
    ```

4.  **Explanation**: This code extracts `title` and `author` from the request body (`req.body`). It performs basic validation to ensure both properties are present, returning a `400 Bad Request` if not. A simple unique `id` is generated by finding the maximum existing `id`, parsing it as an integer, and incrementing it (or starting at 1 if no books exist). The new book object is then added to the `books` array, and the newly created book is returned as a JSON response with a `201 Created` status code.

## Implement PUT Update Book

### Subtask:
Create an Express PUT endpoint (`/books/:id`) that updates an existing book in the in-memory data store. Handle cases where the book is not found or when required fields are missing.

#### Instructions

1.  **Open your `server.js` (or `app.js`) file.**
2.  **Add the following JavaScript code** after your existing POST endpoint. This code defines a new PUT route that listens for requests to `/books/:id`:

    ```javascript
    app.put('/books/:id', (req, res) => {
        const { id } = req.params;
        const { title, author } = req.body;

        let bookIndex = books.findIndex(b => b.id === id);

        if (bookIndex === -1) {
            return res.status(404).send('Book not found');
        }

        if (!title || !author) {
            return res.status(400).send('Title and author are required');
        }

        const updatedBook = { id, title, author };
        books[bookIndex] = updatedBook;
        res.json(updatedBook);
    });
    ```

3.  **Explanation**: This code extracts the `id` from the URL parameters and `title`, `author` from the request body. It uses `findIndex()` to locate the book by its `id`. If the book is not found, a `404 Not Found` response is sent. It also validates that `title` and `author` are provided in the request body, sending a `400 Bad Request` if they are missing. If the book is found and data is valid, the book at the found index is updated with the new details, and the updated book is returned as a JSON response.

## Implement DELETE Book

### Subtask:
Create an Express DELETE endpoint (`/books/:id`) that removes a book from the in-memory data store. Handle cases where the book is not found.

#### Instructions

1.  **Open your `server.js` (or `app.js`) file.**
2.  **Add the following JavaScript code** after your existing PUT endpoint. This code defines a new DELETE route that listens for requests to `/books/:id`:

    ```javascript
    app.delete('/books/:id', (req, res) => {
        const { id } = req.params;

        const initialLength = books.length;
        books = books.filter(book => book.id !== id);

        if (books.length < initialLength) {
            res.status(204).send(); // No Content
        } else {
            res.status(404).send('Book not found');
        }
    });
    ```

3.  **Explanation**: This code extracts the `id` from the URL parameters. It then filters the `books` array, creating a new array that excludes the book with the matching `id`. If the length of the `books` array changes (meaning a book was removed), a `204 No Content` status is returned to indicate successful deletion. If the length does not change, it means no book with the given ID was found, and a `404 Not Found` response is sent.
